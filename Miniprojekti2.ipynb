```python
# Mitä kaikkia data-alan titteleitä dataframesta löytyy?

df_job_titlet = df['job_title'].unique()
df_list = df_job_titlet.tolist()
df_list
```

```python
# Toisen dataframen tittelit. 

df2_job_titlet = df2['Job Title'].unique()
df2_job_titlet

#Junior ja Senior -lisänimet ovat tässä df:ssä yhdistetty titteliin
```

```python
# Jr ja Se -arvojen erittely omaan kolumniin
# Sekä em. arvojen poistaminen job titlestä

df2['Experience_Level'] = df2['Job Title'].str.extract(r'\b(Junior|Senior)\b', expand=False)
df2['Job Title'] = df2['Job Title'].str.replace(r'\b(Junior|Senior)\b', '').str.strip()
df2
```

```python
# Tittelit uudestaan käsittelyn jälkeen

df2_job_titlet = df2['Job Title'].unique()
df2_list = df2_job_titlet.tolist()
df2_list
```

```python
# Dataframeissa experience level on ilmaistu eri tavoin

df['experience_level'].unique()
```

```python
# EN (entry level) / JR, SE, EX --> Junior, Senior, Executive -muotoon

tyokokemus = {
    'JR': 'Junior',
    'EN': 'Junior',
    'SE': 'Senior',
    'EX': 'Executive',
    'MI': 'Mid-Level'
}

df['experience_level'] = df['experience_level'].replace(tyokokemus)
df.head(3)
```

```python
## Dataframeja yhdistävät Job Titlet

def intersection(a, b):
    return list(set(a).intersection(b))

tulos = intersection(df_list, df2_list)
print(tulos)
```

```python
# df ja df2 trimmattuna tulos -muuttujan perusteella

df_data = df.loc[df['job_title'].isin(['Data Analyst', 'Data Scientist', 'Data Engineer', 'Director of Data Science', 'Research Scientist'])]
df_data.head()
```

```python
df2_data = df2.loc[df2['Job Title'].isin(['Data Analyst', 'Data Scientist', 'Data Engineer', 'Director of Data Science', 'Research Scientist'])]
df2_data['Experience_Level'].fillna('Mid-Level', inplace=True)
df2_data.head()
```

```python
## Yhtenäistetään & järjestellään kolumneja

df_data =df_data.rename(columns={"experience_level": "Experience Level", \
                   "job_title":"Job Title", "salary_in_usd":"Salary_In_Usd"}, errors="raise")

df2_data = df2_data.rename(columns={"Experience_Level": "Experience Level"})
```

```python
# Dimensioita varten uusi dataframe jossa vain uniikit rivit

dimcompany = df_data.loc[:, ['company_location', 'company_size']].reset_index(drop=True)
dimcompany = dimcompany.drop_duplicates(subset=['company_location', 'company_size']).reset_index(drop=True)
dimcompany
```

```python
# DimPersonia varten oma dataframe 

dimperson = df2_data.loc[:, ['Gender', 'Education Level', 'Age', 'Years of Experience']].reset_index(drop=True)
dimperson = dimperson.drop_duplicates(subset=['Gender', 'Education Level', 'Age', 'Years of Experience']).reset_index(drop=True)
dimperson
```
